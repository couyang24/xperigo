### 3.3 Data Cleaning: Correcting, Completing, Creating, and Converting

#### 3.3.1 Correcting & Completing

As we can see from Latitude and Longtitude, some numbers are certainly off the scale and better to be removed.

```python
rev_colli = colli[colli.LATITUDE>0]
rev_colli = rev_colli[rev_colli.LATITUDE<40.95]
rev_colli = rev_colli[rev_colli.LONGITUDE>-74.5]
rev_colli = rev_colli[rev_colli.LONGITUDE<-73]
```

```python
display(rev_colli.describe())
print('New Number of rows: ', rev_colli.shape[0])
del(colli)
```

### 4 Interactive Map

```python
lat = rev_colli['LATITUDE'].tolist()
lon = rev_colli['LONGITUDE'].tolist()
locations = list(zip(lat, lon))
```

```python
map1 = folium.Map(location=[40.723204, -73.923018], tiles='OpenStreetMap', zoom_start=11.5)
FastMarkerCluster(data=locations).add_to(map1)
map1.save('map1.html')
```
